```python
import numpy as np
#please consider a1 as x and a2 as y thank you 
a1 = np.arange(1,8)
a2 = np.array([2,3,5,4,6,8,9])
covMatrix = np.cov(a1,a2)
covMatrix
a1Mean = (a1.mean())
a1Mean
n=7
yMean = a2.mean()
diff1 = a1 - a1Mean
diff2 = a2- yMean

diff1
prodXY = (diff1 * diff2).round(2)
prodXX = (diff1 * diff1).round(2)
prodYY = (diff2 * diff2).round(2)
prodYX = (diff2 * diff1).round(2)
covXY = (prodXY.sum())/(n-1)
covXX = (prodXX.sum())/(n-1)
covYY = (prodYY.sum())/(n-1)
covYX = (prodYX.sum())/(n-1)
print("the covariance of X and Y  is", covXY)
print("the covariance of X and X  is", covXX)
print("the covariance of Y and Y  is", covYY)
print("the covariance of Y and X  is", covYX)
finalCov = np.array([covXX,covXY,covYX,covYY]).reshape(2,2)
finalCov
# print("the before covariance is ", covMatrix) 

```
